```python
import pandas as pd
import glob
```

```python
# Load and combine all raw CSV files
csv_files = glob.glob("../data/raw_data/*.csv")
print(csv_files)
df_list = [pd.read_csv(f) for f in csv_files]
df = pd.concat(df_list, ignore_index=True)

# Clean and organize dataset
df = df.drop_duplicates()
df = df.sort_values(by='datetimeUtc')
df = df.reset_index(drop=True)

# Save combined dataset
df.to_csv("../data/processed_data/NYC_combined_data.csv", index=False)

print("Combined CSV created successfully!")
```
