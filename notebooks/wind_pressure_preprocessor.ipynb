```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw_data/wind_data.txt")
df
```

```python
# Clean and convert wind/pressure fields
df['drct'] = pd.to_numeric(df['drct'], errors='coerce')
df['alti'] = pd.to_numeric(df['alti'], errors='coerce')

# Convert altimeter to pressure (mb) and round
df['slp'] = df['alti'] * 33.8639
df['slp'] = df['slp'].round(2)

# Rename and restructure columns
df = df.rename(columns={
    'valid': 'datetimeUtc',
    'sknt': 'wind_speed',
    'drct': 'wind_direction',
    'slp': 'pressure'
})

df = df[['station', 'datetimeUtc', 'wind_speed', 'wind_direction', 'pressure']]

# Save processed dataset
df.to_csv("../data/processed_data/NYC_wind_pressure.csv", index=False)
```
